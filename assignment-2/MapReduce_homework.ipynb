#### MapReduce

Use MapReduce to calculate the average housing price in each zip code and filter out records in which gross sqft is less than 500 and sale price is less than 100000. Round the average housing price to 2 decimals, and sort the output by housing price ascending.

```python
# !wget -O housingSalseSample.csv https://github.com/CUSP2022ADS/Data/raw/main/housingSalseSample.csv
import pandas as pd
df = pd.read_csv('housingSalseSample.csv')
df
```

```python
# code here
from functools import reduce

def priceByZipcode(result,record):
    zipcode, price = record[0],record[1]
    result[zipcode] = result.get(zipcode,[]) + [price]
    return result

# reduce(priceByZipcode,df[['ZIP CODE', 'SALE PRICE']].values,{})
def averagePrice(record):
    zipcode, prices = record[0], record[1]
    return [zipcode, round(sum(prices)/len(prices),2)]

output = list(filter(lambda x: x[2]<500 and x[1]<100000, df[['ZIP CODE', 'SALE PRICE', 'GROSS SQUARE FEET']].values))
output = list(map(averagePrice,list(reduce(priceByZipcode, output, {}).items())))
output = sorted(output, key=lambda x: x[1])
output
```

```python
'''grading function
please pass your output into this function and upload the generated csv file 
together with your notebook to your GitHub repo'''

def grading(output):
    import pandas as pd
    pd.DataFrame(output,columns=['zipcode','price']).to_csv('output1.csv',index=False)

grading(output)
```

Apply the chunk and MapReduce method to sum up each type's value in the synthetic dataset—only keep records which $Select$ value is 1, and read 1000 lines in each chunk. Use multiprocessing package to deploy the task to multi-cores. The output should only have two columns: type and total value.

Hint: filter function is not available in multiprocessing, implement an if-else function in the map or reduce step.

```python
# from multiprocessing import Pool # this gets error: AttributeError: can't get attribute 'returnInfo' on <module '__main__' (built-in)>
from multiprocess import Pool # this works. requirement: pip install multiprocess
from functools import reduce
import csv

def returnInfo(record): #extract necessary fields from a data record (raw)
    if record['Select'] == '1':
        return(record['Type'], record['Value'])
    else:
        return(record['Type'], 0)

def sumValue(result,record): #sum up each type's value
    t, v = record[0], record[1]

    result[t] = result.get(t, 0) + float(v)

    return result

with open('ADSSession2Task2Data.csv', 'r') as fi:
    reader = csv.DictReader(fi)
    chunk = []
    result = []
    # read 1000 lines in each chunk
    chunksize = 1000
    pool = Pool(8)
    for i, line in enumerate(reader):
        if (i % chunksize == 0 and i > 0):
            # MapReduce function for rows in one chunk
            output = list(reduce(sumValue, pool.map(returnInfo, chunk), {}).items())
            result += output
            del chunk[:]  # or: chunk = []
                 
        # add lines to chunk, and when the index
        # is 1000, 2000, etc, stop adding new lines
        # and start MapReduce functions
        chunk.append(line)
        
    # MapReduce function for the last chunk
    output = list(reduce(sumValue, pool.map(returnInfo, chunk), {}).items())
    result += output
    pool.close()
```

```python
output = list(reduce(sumValue,result,{}).items())[:-1]
output
```
